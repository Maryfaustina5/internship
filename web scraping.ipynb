```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://en.wikipedia.org/wiki/Main_Page"
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")
h = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])
hr = [tag.get_text() for tag in h]
df = pd.DataFrame(hr)
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL for the ICC ODI Men's Team Rankings
team_ranking_url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"

# URL for the ICC ODI Men's Batsmen Rankings
batsmen_ranking_url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"

# URL for the ICC ODI Men's Bowlers Rankings
bowlers_ranking_url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"

# Function to scrape ranking data and create a DataFrame
def scrape_and_create_dataframe(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        ranking_table = soup.find('table', class_='table')
        data = []

        for row in ranking_table.find_all('tr')[1:11]:  # Extract the top 10 records
            cols = row.find_all('td')
            record = [col.text.strip() for col in cols]
            data.append(record)

        headers = ['Rank', 'Team', 'Matches', 'Points', 'Rating']
        df = pd.DataFrame(data, columns=headers)
        return df
# Scrape and create DataFrames for Team Rankings, Batsmen Rankings, and Bowlers Rankings
team_df = scrape_and_create_dataframe(team_ranking_url)
batsmen_df = scrape_and_create_dataframe(batsmen_ranking_url)
bowlers_df = scrape_and_create_dataframe(bowlers_ranking_url)

# Display the DataFrames
print("Top 10 ODI Teams:")
print(team_df)

print("\nTop 10 ODI Batsmen:")
print(batsmen_df)

print("\nTop 10 ODI Bowlers:")
print(bowlers_df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the CNBC World News page
url = "https://www.cnbc.com/world/?region=world"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Initialize empty lists to store data
    headlines = []
    times = []
    news_links = []

    # Find and extract news details
    news_articles = soup.find_all("div", class_="Card-textWrapper")
    for article in news_articles:
        headline = article.find("a", class_="Card-title").text.strip()
        time = article.find("time")['datetime']
        news_link = article.find("a", class_="Card-title")['href']
        headlines.append(headline)
        times.append(time)
        news_links.append(news_link)
# Create a DataFrame
    news_df = pd.DataFrame({
        'Headline': headlines,
        'Time': times,
        'News Link': news_links
    })

    # Display the DataFrame
    print(news_df)

else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page with most downloaded AI articles
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Initialize empty lists to store data
    titles = []
    authors = []
    published_dates = []
    paper_urls = []

    # Find and extract article details
    articles = soup.find_all("li", class_="pod-listing")
    for article in articles:
        title = article.find("a", class_="pod-listing-title").text.strip()
        author = article.find("a", class_="pod-listing-author-name").text.strip()
        date = article.find("div", class_="pod-listing-body").find("p").text.strip()
        url = article.find("a", class_="pod-listing-title")['href']
        titles.append(title)
        authors.append(author)
        published_dates.append(date)
        paper_urls.append(url)

    # Create a DataFrame
    article_df = pd.DataFrame({
        'Paper Title': titles,
        'Authors': authors,
        'Published Date': published_dates,
        'Paper URL': paper_urls
    })
    
     # Display the DataFrame
    print(article_df)

else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```
